```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
data = pd.read_csv("data/NBA team-rating.csv")
data
```

```python
data['path'] = "data/NBAlogo/"+data.Team+".png"
data
```

```python
import matplotlib.font_manager
fm = matplotlib.font_manager
fm._get_fontconfig_fonts.cache_clear()
path = 'C:\\Users\\pc\\AppData\\Local\\Microsoft\\Windows\\Fonts\\zil-1.ttf'
prop = matplotlib.font_manager.FontProperties(fname=path)
plt.rcParams['font.family'] = prop.get_name()
fig, ax = plt.subplots(figsize=(6, 4), dpi=120)
ax.scatter(data.ORtg,data.DRtg)
def getImage(path):
    return OffsetImage(plt.imread(path), zoom=.05, alpha = 1)
for index, row in data.iterrows():
    ab = AnnotationBbox(getImage(row['path']), (row['ORtg'], row['DRtg']), frameon=False)
    ax.add_artist(ab)
ax.hlines(data['ORtg'].mean(), data['ORtg'].min(), data['ORtg'].max(), color='#c2c1c0')
ax.vlines(data['DRtg'].mean(), data['DRtg'].min(), data['DRtg'].max(), color='#c2c1c0')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('#ccc8c8')
ax.spines['bottom'].set_color('#ccc8c8')
ax.set_xlabel("ORtg")
ax.set_ylabel("DRtg")
ax.set_title("NBA Teams Performance")
```
